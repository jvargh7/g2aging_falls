```python
import openpyxl
import msoffcrypto
import io
import os
import pandas as pd
import re
```

```python
# Load the workbook
path_merck_proposal = 'C:/Cloud/OneDrive - Emory University/Proposals/Merck Atherosclerosis and Diabetes'
# https://stackoverflow.com/questions/19450837/how-to-open-a-password-protected-excel-file-using-python
def load_cosmos_excel(file_name):
    decrypted_workbook = io.BytesIO()

    with open(path_merck_proposal + '/working/' + file_name, 'rb') as file:
        office_file = msoffcrypto.OfficeFile(file)
        office_file.load_key(password='Cosmos@2023')
        office_file.decrypt(decrypted_workbook)
        df = pd.read_excel(decrypted_workbook,skiprows=3,names=['county','counts'])
        # Add new column to df with file name
        df['file_name'] = file_name
        # Add state column from file_name
        df['state'] = re.split('Bottom|Top',file_name)[0]
        


    return df
```

```python
# https://stackoverflow.com/questions/28669482/appending-pandas-dataframes-generated-in-a-for-loop
all_dfs = []
# https://stackoverflow.com/questions/3207219/how-do-i-list-all-files-of-a-directory
list_of_files = os.listdir(path_merck_proposal + '/working')
for f in list_of_files:
    if f.endswith('.xlsx'):
        dataset = load_cosmos_excel(f)
        all_dfs.append(dataset)

all_dfs = pd.concat(all_dfs)
all_dfs.tail()

all_dfs.to_csv(path_merck_proposal + '/working/merged_cosmos.csv',index=False)
```
